# Univariate Analysis

```python
# Univariate Analysis
plt.style.use('seaborn')
fig, ax = plt.subplots(1,2, figsize=(16,6))

release_period = df['Release Period']
genre = df['Genre']


sns.countplot(x = release_period, data = df, ax = ax[0], palette = 'Set2')
sns.countplot(x = genre, data = df, ax = ax[1], palette = 'Set2')

plt.xticks(rotation = 90)
plt.tight_layout()
```

```python
fig, ax = plt.subplots(1,2, figsize=(16,6))

whether_remake = df['Whether Remake']
whether_franchise = df['Whether Franchise']


sns.countplot(x = whether_remake, data = df, ax = ax[0], palette = 'Set2')
sns.countplot(x = whether_franchise, data = df, ax = ax[1], palette = 'Set2')

plt.tight_layout()
```

```python
fig, ax = plt.subplots(1,2, figsize=(16,6))

new_actor = df['New Actor']
new_music_director = df['New Music Director']


sns.countplot(x = new_actor, data = df, ax = ax[0], palette = 'Set2')
sns.countplot(x = new_music_director, data = df, ax = ax[1], palette = 'Set2')

plt.tight_layout()
```

```python
df['Lead Star'].value_counts().sort_values(ascending = False).head(10).plot(kind = 'bar')
plt.title('Lead Star Plot')
plt.tight_layout()
```

```python
df['Director'].value_counts().sort_values(ascending = False).head(10).plot(kind = 'bar')
plt.title('Director Plot')
```

```python
df['Music Director'].value_counts().sort_values(ascending = False).head(10).plot(kind = 'bar')
plt.title('Music Director Plot')
plt.tight_layout()
```

# Multivariate Analysis

```python
# Multivariate analysis 

df.groupby(['Lead Star'])['Whether Remake'].count().sort_values(ascending = False).head(10).plot(kind = 'bar')
plt.title('Top remake actors')
plt.tight_layout()
```

```python
a = df.groupby(['Lead Star','Director'])['Whether Remake'].count().sort_values(ascending = False).head(10).reset_index()
a['actor and director'] = a['Lead Star'] + a['Director']
a.drop(columns = ['Lead Star'], inplace = True)
a.drop(columns = ['Director'], inplace = True)
sns.barplot(x = a['actor and director'], y = a['Whether Remake'], data = a)

plt.xticks(rotation = 90)
plt.title('Number of duplicate movies made by pair of actor and director')
# plt.tight_layout()
```

```python
df.groupby(['Lead Star'])['Whether Franchise'].count().sort_values(ascending = False).head(10).plot(kind = 'bar')
plt.tight_layout()
plt.title('Actor whose Done Number of Movie Franchise')
plt.tight_layout()
```

```python
df['Profit'] = df['Revenue(INR)'] - df['Budget(INR)']

a = df.groupby(['Director'])['Revenue(INR)'].value_counts().sort_values(ascending = False).head(10).reset_index()
a.sort_values(by = 'Revenue(INR)')
a.drop(columns = ['count'], inplace = True)

sns.lineplot(x = a['Director'], y = a['Revenue(INR)'], data = a)
plt.xticks(rotation = 90)
plt.title('Most Profitable Director')
plt.tight_layout()
```

```python
# Creating a function that returns all kind of movies for every actors
def actors(df, actor_name):
    lead_actor_dataset = df[df['Lead Star'] == actor_name]
    sns.lineplot(x = lead_actor_dataset['Movie Name'], y = lead_actor_dataset['Revenue(INR)'], data = lead_actor_dataset)
    plt.xticks(rotation = 90)
    plt.title(actor_name + ' Revenue Movies')
    plt.tight_layout()
```

```python
actors(df,'Ajay Devgn')
```
